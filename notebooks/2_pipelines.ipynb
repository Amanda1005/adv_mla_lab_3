```python
# === [Step 0] Import 所需模組 ===
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import SGDRegressor
from sklearn.compose import ColumnTransformer

# === [Step 1] 讀取資料與定義 X, y ===
df = pd.read_csv("../../work/data/processed/X_train.csv")
df_cleaned = df.copy()

X = df_cleaned.drop(columns=["pts"])  # 特徵
y = df_cleaned["pts"]                 # 目標

# === [Step 2] 數值欄位名稱（全部都是 numeric）===
num_cols = X.columns.tolist()

# === [Step 3] 數值欄位的處理流程（補值 + 標準化）===
num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

# === [Step 4] 建立前處理器 ColumnTransformer（目前只有 numeric）===
preprocessor = ColumnTransformer(transformers=[
    ("num", num_transformer, num_cols)
])

# === [Step 5] 建立完整 Pipeline，使用 SGDRegressor 進行回歸 ===
sgd_pipe = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("sgd", SGDRegressor(loss="squared_error", penalty="l2", max_iter=1000, random_state=42))
])

# === [Step 6] 模型訓練 ===
sgd_pipe.fit(X, y)
```
